# 실습 #3: 워드 클라우드

1) 인터넷에서 좋은 글(영문으로 된 연설문, 혹은 기사, 책 내용 등)을 찾아, txt로 저장하고 읽어와 워드 클라우드를 만들어 보세요.

- 코드는 예제파일을 참조하세요.
- 불필요한 단어/조사는 제거하세요.
- 글의 내용을 한눈에 파악할 수 있는 멋진 워드클라우드를 만들어 보세요.

```python
# 파일 읽기
file = open('인연.txt', 'r', encoding='UTF-8')
text = file.read()
file.close()

# 단어 단위로 자르기
wordList = []

# 한 글자, 마참표 있는 단어 제거
for x in text.split():
    if '.' in x:
        pass
    elif len(x) == 1:
        pass
    else:
        wordList.append(x)
```

```python
# 중복 단어 제거
worduniq = set(wordList)

# 딕셔너리 선언
wordCount = {}

# 단어별 개수 저장
for w in worduniq:
    wordCount[w] = wordList.count(w)

# 조사 제외
del_word = ['그리고', '있는', '이런', '그는', '이런', '아니', '하고', '같은',
            '하다가', '있었고', '내가', '나는', '그러나', '것은', '그동안',
            '하게', '같이', '것도', '것을','어쩌다']
for w in del_word:
    if w in wordCount:
        del wordCount[w]
```

```python
# 라이브러리 불러오기
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import numpy as np
from PIL import Image
from wordcloud import ImageColorGenerator

# 이미지 불러오기
masking_image = np.array(Image.open('인연.jpg'))

# 워드 클라우드 만들기
wordcloud = WordCloud(font_path = 'C:/Windows/fonts/HMKMRHD.TTF',
                      width=2000,
                      height=1000,
                      mask=masking_image,
                      max_font_size=100,
                      background_color='white').generate_from_frequencies(wordCount)

# 표시하기
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
